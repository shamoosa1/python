```python
## this web scraping project is a tool to sort books on a website based on their star reviews (1 - 5 stars)

import requests
import bs4

# the book website we will be scraping
base_url = "http://books.toscrape.com/catalogue/page-{}.html"

# create an empty list of book titles that will be stored
five_star_books = []
four_star_books = []
three_star_books = []
two_star_books = []
one_star_books = []

## this function will sort all books of whichever range of pages you specify
def organizeByPageRange(start, finish):

# creating a soup object to inspect the page
# select all books by locating the product_pod class

    for page_num in range(start,finish): # testing for all pages
    
        scrape_url = base_url.format(page_num)
        result = requests.get(scrape_url)

        soup = bs4.BeautifulSoup(result.text, "lxml")

        books = soup.select(".product_pod")
        
        for book in books:

            book_title = book.select("a")[1]["title"]

            if len(book.select(".star-rating.Five")) != 0:
                five_star_books.append(book_title)

            elif len(book.select(".star-rating.Four")) != 0:
                four_star_books.append(book_title)

            elif len(book.select(".star-rating.Three")) != 0:
                three_star_books.append(book_title)

            elif len(book.select(".star-rating.Two")) != 0:
                two_star_books.append(book_title)

            else:
                one_star_books.append(book_title) 


## this function should only be used to sort a specified page number
def organizeByPage(page_num): 
    

    scrape_url = base_url.format(page_num)
    result = requests.get(scrape_url)

# creating a soup object to inspect the page
    soup = bs4.BeautifulSoup(result.text, "lxml")

# select all books by locating the product_pod class
    books = soup.select(".product_pod")

    for book in books:

        book_title = book.select("a")[1]["title"]

        if len(book.select(".star-rating.Five")) != 0:
            five_star_books.append(book_title)

        elif len(book.select(".star-rating.Four")) != 0:
                four_star_books.append(book_title)

        elif len(book.select(".star-rating.Three")) != 0:
                three_star_books.append(book_title)

        elif len(book.select(".star-rating.Two")) != 0:
                two_star_books.append(book_title)
                
        else:
                one_star_books.append(book_title) 


```
